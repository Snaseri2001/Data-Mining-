# Lower case
First we convert the text data to lower case. One of its benefit is that we can calculate the frequency of each word as simple as possible.

```python

def remove_html(Series):
    for item in range(Series.size):
        text = str(Series[item])
        html_pattern = re.compile('<.*?>')
        Series.replace(Series[item] , html_pattern.sub(r'', text), inplace = True)
    print(Series)
remove_html(University_data["academic_req"])
```

# Removal of Punctuations
One another common text preprocessing technique is to remove the punctuations from the text data.for example "[this is home !!] " is better to be "this is home "

```python
# First we have to fill the NAN value in thease 2 series
University_data["structure"].fillna("no value", inplace =True)
University_data["city"].fillna("no value" , inplace = True)
```

```python
PUNCT_TO_REMOVE = string.punctuation
def remove_punctuation(text):
    return text.translate(str.maketrans('', '', PUNCT_TO_REMOVE))

University_data["structure_wo_punct"] = University_data["structure"].apply(lambda text: remove_punctuation(text))
University_data["city_wo_punct"] = University_data["city"].apply(lambda text: remove_punctuation(text))
```

```python
# now we have to remove "city" and "structure" columns
University_data.drop( University_data.columns[[ 0 , 8 , 10 ,11 ]] , axis = 1  ,  inplace=True)
University_data
```

# Text Cleaning
Here we have to feel in the Nan elements. As we see we have lots of Nan value "ielts_score" serie so must design a policy for this column. we use the avarage of the ielts_scores of the same university for Nan fields.

```python

for item in range(University_data["ielts_score"].size) : 
    if pd.isna(University_data["ielts_score"][item]) :
        
        avarage_score_of_the_same_university =University_data.loc[University_data["university_name"] == University_data["university_name"][item] , "ielts_score"].mean()
        University_data["ielts_score"].iloc[ item] = avarage_score_of_the_same_university 
        
University_data
```

```python
# here we are going to remove the remained data with Non "ielts_score" field.
size_of_dataFrame = University_data["ielts_score"].size
for item in range(size_of_dataFrame) : 
    if pd.isna(University_data["ielts_score"][item]) :
        University_data.drop(item , inplace=True)
University_data
```

```python
# here we remove the data with null "rank" feild 
size_of_dataFrame = University_data["university_rank"].size
for item in range(size_of_dataFrame) : 
    if pd.isna(University_data["university_rank"][item]) :
        University_data.drop(item , inplace=True)
University_data
```

# Remove stopwards
here we have to remove stopwards like "the" , "a" , etc . these words dont provide valuable information .

```python
# if we want use nltk we have to download the data using :
# nltk.download('stopwords')
# the link below can guide us to download 
#  https://www.projectpro.io/recipes/download-nltk-library
```

```python
STOPWORDS = set(stopwords.words('english'))
def remove_stopwords(text):
    """custom function to remove the stopwords"""
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])
```

```python
#  we have to remove stopwards in "program_name" , "academic_req" , "structure_wo_punct" , "city_wo_punct"
University_data["program_name"] = University_data["program_name"].apply(lambda text: remove_stopwords(text))
University_data["academic_req"] = University_data["academic_req"].apply(lambda text: remove_stopwords(text))
University_data["structure_wo_punct"] = University_data["structure_wo_punct"].apply(lambda text: remove_stopwords(text))
University_data["city_wo_punct"] = University_data["city_wo_punct"].apply(lambda text: remove_stopwords(text))
University_data
```
