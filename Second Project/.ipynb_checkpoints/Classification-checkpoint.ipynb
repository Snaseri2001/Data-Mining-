# Decision Tree
Firstly we are going to predict "SHOOTING" feature via decision tree.


```python
Crime = pd.read_csv("data_Category.csv")
X1 = Crime[["DAY_OF_WEEK"  , "YEAR" , "MONTH" , "HOUR" , "X_Location" , "Y_Location" , "OFFENSE_CODE_GROUP" , "STREET" ]]
Y1 = Crime["OFFENSE_CODE_CATEGORY"]
y=Y1.dropna()
X1_train, X1_test, Y1_train, Y1_test = train_test_split( X1, y, test_size=0.1, random_state=0)

```

```python
clf = tree.DecisionTreeClassifier()
clf.fit(X1_train , Y1_train)
```

```python
clf.score(X1_test , Y1_test)
```

# Multy Layer Perceptron (MLP)


```python
Crime =  pd.read_csv("data_Category.csv")
X2  = Crime[["DAY_OF_WEEK"  , "YEAR" , "MONTH" , "HOUR" , "X_Location" , "Y_Location" ,"SHOOTING" , "STREET" ]]
Y2 = Crime["OFFENSE_CODE_CATEGORY"]

# here we have to scale the training data
scaled = MinMaxScaler()
scaled.fit(X2)
help = scaled.transform(X2)
X2 = pd.DataFrame(help, columns=X2.columns)

X2_train, X2_test, Y2_train, Y2_test = train_test_split( X2, Y2, test_size=0.1, random_state=0)
```

```python
# create the model


nn_model = ks.models.Sequential()
nn_model.add(ks.layers.Flatten(input_shape = ( 8, )))
nn_model.add(ks.layers.Dense( 60 , activation = 'relu'))
nn_model.add(ks.layers.Dense( 40, activation = 'relu'))
nn_model.add(ks.layers.Dense(61 , activation = "softmax"))
nn_model.compile(optimizer= "Adam" , loss='sparse_categorical_crossentropy',metrics=['accuracy'])
nn_model.fit(X2_train , Y2_train ,epochs=400)
y_pred = nn_model.predict(X2_test)
```

```python
Crime =  pd.read_csv("data_Category.csv")
X  = Crime[["DAY_OF_WEEK"  , "YEAR" , "MONTH" , "HOUR" , "X_Location" , "Y_Location" ,"SHOOTING" , "STREET" ]]
y = Crime["OFFENSE_CODE_CATEGORY"]
scaled = MinMaxScaler()
scaled.fit(X)
help = scaled.transform(X)
X2 = pd.DataFrame(help, columns=X.columns)
X_train, X_test, y_train, y_test = train_test_split( X2, y, test_size=0.1, random_state=0)
```

```python
from sklearn.svm import SVC
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score




svm_model = SVC()

svm_model.fit(X_train, y_train)

y_pred = svm_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy  : {:.2f}".format(accuracy))
```

```python
from keras.models import Sequential
from keras.layers import Dense
from sklearn.metrics import mean_squared_error
# ساختن یک مدل شبکه عصبی
model = Sequential()
model.add(Dense(256, activation='relu', input_dim=8))
model.add(Dense(64, activation='relu'))
model.add(Dense(61, activation='sigmoid'))

# کامپایل مدل
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# آموزش مدل با داده‌های آموزشی

model.fit(X_train, y_train, epochs=200, batch_size=32)

# پیش‌بینی برچسب‌های داده‌های آزمون
y_pred = model.predict(X_test)


```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.multioutput import MultiOutputClassifier
randomForest = RandomForestClassifier(max_depth=3, criterion='gini',min_samples_leaf=30, n_estimators=120, n_jobs=-1, random_state=1122)

randomForest.fit(X_train, y_train)

print('Accuracy on test data: {:.1f}%'.format(accuracy_score(y_test, randomForest.predict(X_test))*100))
print('Accuracy on train data: {:.1f}%'.format(accuracy_score(y_train, randomForest.predict(X_train))*100))

print('F1 score on test data: {:.1f}%'.format(accuracy_score(y_test, randomForest.predict(X_test))*100))
print('F1 score on train data: {:.1f}%'.format(accuracy_score(y_train, randomForest.predict(X_train))*100))
```
