## Featuer Scaling using `Standardization`
### `sklearn` has `StandardScaler` for scaling data

```
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

scaler.fit(X)
scaler.transform(new_X)
scaler.fit_transform(X)
```

### `fit` method :
Gets `X` as input dataset then calculates and saves `std` and `mean`. It DOES NOT scale the data. ONLY calculate `std` and `mean`.

### `transform` method :
Gets `X` as input dataset then SCALE it using calculated `std` and `mean` from `fit` method.

### `fit_transform` method :
First call `fit` then call `transform` method.

````python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()

X = pd.DataFrame(std_scaler.fit_transform(X), columns = dataset.iloc[:, [3,4,5,6,9,13]].columns)
X
````

````python
# elbow method 
from sklearn.cluster import KMeans
cost = []

for i in range(1,21):
    kmeans = KMeans(n_clusters= i, init='k-means++', random_state=0, n_init=5)
    kmeans.fit(X)
    cost.append(kmeans.inertia_)
````

````python
#Visualizing the ELBOW method to get the optimal value of K 
plt.plot(range(1,21), cost)
plt.title('The Elbow Method')
plt.xlabel('no of clusters')
plt.ylabel('cost')
plt.show()
````

````python

# finally we got that k=5

kmeans_cluster = KMeans(n_clusters= 8, init='k-means++', random_state=0)
kmeans_cluster_pred = kmeans_cluster.fit_predict(X)
````

## DBSCAN Model

````python
sh_dbscna=shooting.drop(columns=['labels','cluster predict by kmodes'],axis=1)
sh_dbscna
````

````python

from sklearn.decomposition import PCA


# Perform dimensionality reduction using PCA
pca = PCA(n_components=2)
sh_pca = pca.fit_transform(sh_dbscna)

plt.scatter(sh_pca[:, 0], sh_pca[:, 1], c='blue')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Data Points')
plt.show()
````

````python
import numpy as np
from sklearn.metrics.pairwise import euclidean_distances
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

def ddscan(x, eps, min_pts):
    # Calculate pairwise distances
    distances = euclidean_distances(x)
    
    # Initialize visited and cluster labels
    visited = np.zeros(x.shape[0], dtype=bool)
    labels = np.zeros(x.shape[0], dtype=int)
    cluster_id = 0
    
    # Iterate through all data points
    for i in range(x.shape[0]):
        if not visited[i]:
            visited[i] = True
            
            # Find neighbors within epsilon distance
            neighbors = np.where(distances[i] <= eps)[0]
            
            if len(neighbors) < min_pts:
                labels[i] = -1  # Mark as noise point
            else:
                cluster_id += 1
                labels[i] = cluster_id
                
                # Expand the cluster
                expand_cluster(x, distances, visited, labels, i, neighbors, cluster_id, eps, min_pts)
    
    return labels

def expand_cluster(x, distances, visited, labels, index, neighbors, cluster_id, eps, min_pts):
    i = 0
    while i < len(neighbors):
        neighbor = neighbors[i]
        if not visited[neighbor]:
            visited[neighbor] = True
            
            # Find neighbors of the neighbor
            neighbor_neighbors = np.where(distances[neighbor] <= eps)[0]
            
            if len(neighbor_neighbors) >= min_pts:
                neighbors = np.concatenate((neighbors, neighbor_neighbors))
            
            # Assign neighbor to the cluster
            if labels[neighbor] == 0 or labels[neighbor] == -1:
                labels[neighbor] = cluster_id
        
        i += 1

# Example usage:



````

````python
eps = 3
min_pts = 2

# Run DDScan clustering
labels = ddscan(sh_dbscna, eps, min_pts)
print("Cluster labels:", labels)
````
